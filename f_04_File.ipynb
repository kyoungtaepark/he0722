```python
import numpy as np
from numpy.random import rand
import csv
```

##  숫자 데이터  파일 생성

```python
n=5
lines =10
with open('test_1.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    for i in range(lines):
        row = []
        for i in range(n):
            row.append(round(rand(),3)) # 임의의 랜덤 넘버 생성 (소수점 3자리)
        writer.writerow(row)
```

## 문자 데이터 쓰기

```python
# 쓰기 전용으로 파일 열기 (write)
# 파일이 없으면 새로 만든다
f = open("test.csv", "w")
f.write("a,b,c,d,e \n")
f.write("1,2,3,4,5 \n")
f.close()
```

```python
# 파일 내용 보기
with open('test.csv') as f:
    contents = f.read()
    print(contents)
```

```python
# 파일 내용 일부 읽기
with open('test.csv') as f:
    contents = f.read(30)
    print(contents)
```

### with 사용
- 파일을 열 때 with를 사용하면, 해당 블록을 나갈 때 열린 파일을 닫는다
- 즉, close 호출을 잊어버려도 안전하게 파일을 닫는다

```python
# 운영체제 명령으로 보기
# !cat test.csv  # 리눅스나 맥에서
# !type test.csv # 윈도우에서
```

### 파일 오픈 옵션
'r'	읽기 (디폴트)  
'w'	쓰기 (파일이 없었으면 새로 생성하며 이미 있었으면 덮어쓴다)  
'a'	맨 뒤에 추가 (appending), 파일이 없었으면 새로 생성  
't'	텍스트 모드로 오픈 (디폴트)  
'b'	바이너리 모드로 오픈  
'+'	읽기 쓰기 모두 가능  

```python
# 기존 파일에 추가하는 옵션으로 파일 열기 (append)
with open("test.csv", "a") as f:
    f.write("6,7,8,9,10 \n")
    f.write("11,12,13,14,15 \n")
    f.write("16,17,18,19,20 \n")
```

## zip 파일 만들기

```python
import zipfile
with zipfile.ZipFile('output.zip', 'w') as myzip:
    myzip.write('output.csv')
```

### zip 파일 풀기

```python
with zipfile.ZipFile("output.zip", 'r') as zf: 
    print(zf.namelist()) # 포한된 파일명 보기
    zf.extractall() # 하위 폴더 지정 가능
```

## 현재 폴더 위치 보기

## 폴더의 파일 목록 얻기
- glob() 사용
- listdir() 사용

### glob()를 사용

```python
# 파일 목록 얻기
from glob import glob
sorted(glob('f_*'))
```

### listdir()를 사용

```python
# 파일과 폴더 목록 얻기
file_list = os.listdir()
sorted(file_list)
```

```python
# 파일의 전체 경로, full path 만들기
cur_dir = os.getcwd()
for file_name in file_list:
    full_path = os.path.join(cur_dir, file_name)
    print(full_path)
```

### 폴더 찾기

```python
for file_name in file_list:
    if os.path.isdir(file_name):
        print("폴더:", file_name)
```

```python
!ls d*
```

```python
# 폴더가 없으면 새로 만들기
if not os.path.exists("./data"):
    os.mkdir("./data")
```

```python
# (참고) 폴더가 있으면 "지우고" 새로 만들기
if os.path.exists("./data"):
    os.rmdir("./data")
    os.mkdir("./data")
```
